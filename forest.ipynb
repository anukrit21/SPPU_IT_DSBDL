```python
import pandas as pd
import swifter
```

```python
df = pd.read_csv('forestfires.csv')
df
```

```python
#define the map function
def map_function(row):
    result = row['X'] * row['wind']
    return result
```

```python
#apply the map function to each row in parallel using swifter
df['map_result'] = df.swifter.apply(map_function,axis=1)
df
```

```python
#Perform reduce operation
reduce_result = df['map_result'].sum()
reduce_result
```

```python
#calculate the descriptive statistics of numeric columns
des_stats = df.describe()
des_stats
```

```python
#group the data
group_data = df.groupby('month').agg({'area':['sum','mean','max']})
group_data
```

```python
#plot the graph
import matplotlib.pyplot as plt
df['month'].value_counts().plot(kind='bar')
plt.xlabel('Month')
plt.ylabel('No.of fires')
plt.title('No. of fires per month')
plt.show()
```
